```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets

iris = datasets.load_iris()
X = iris.data
y = iris.target
X = X[y<2, :2]
y = y[y<2]
```

```python
plt.scatter(X[y==0, 0], X[y==0, 1], color="green")
plt.scatter(X[y==1, 0], X[y==1, 1], color="red")
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=500)
```

```python
from LogisticRegression import LogisticRegression
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)
```

```python
log_reg.score(X_test, y_test)
```

```python
log_reg.coef_
```

```python
log_reg.interception_
```

```python
def x2(x1):
    return (-log_reg.coef_[0] * x1 - log_reg.interception_) / log_reg.coef_[1]
```

```python
x1_plot = np.linspace(4, 8, 1000)
x2_plot = x2(x1_plot)
```

```python
plt.scatter(X[y==0, 0], X[y==0, 1], color="green")
plt.scatter(X[y==1, 0], X[y==1, 1], color="red")
plt.plot(x1_plot, x2_plot)
```

```python
plt.scatter(X_test[y_test==0, 0], X_test[y_test==0, 1], color="green")
plt.scatter(X_test[y_test==1, 0], X_test[y_test==1, 1], color="red")
plt.plot(x1_plot, x2_plot)
```
